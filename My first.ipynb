```python
import numpy as np

# Function to calculate total and average
def calculate_scores(scores):
    total = np.sum(scores)  # NumPy function to calculate total
    average = np.mean(scores)  # NumPy function to calculate average
    return total, average

# Create an empty list to store student data
students_data = []

# Get number of students
num_students = int(input("Enter the number of students: "))

# Loop over the number of students
for _ in range(num_students):
    # Get student's name
    name = input("Enter student's name: ")

    # Get scores for 5 subjects
    scores = []
    for i in range(1, 6):
        while True:
            try:
                score = float(input(f"Enter score for subject {i}: "))
                if score < 0 or score > 100:
                    print("Score must be between 0 and 100. Try again.")
                    continue
                scores.append(score)

                break
            except ValueError:
                print("Invalid input! Please enter a numeric value.")

    # Calculate total and average
    total, average = calculate_scores(scores)

    # Check if the student passed or failed
    result = "Pass" if average >= 50 else "Fail"

    # Store the data in the list
    students_data.append((name, scores, total, average, result))

# Print the results for all students
print("\n--- Student Results ---")
for student in students_data:
    name, scores, total, average, result = student
    print(f"Name: {name}")
    print(f"Scores: {scores}")
    print(f"Total: {total}")
    print(f"Average: {average}")
    print(f"Result: {result}")
    print("------------------------")
```

```python
import streamlit as st
import numpy as np

def calculate_scores(scores):
    """Calculates total and average scores."""
    total = np.sum(scores)
    average = np.mean(scores)
    return total, average

# --- Streamlit UI ---
st.title("Student Grading System")

name = st.text_input("Student Name:")
subject1 = st.number_input("Score for Subject 1:", min_value=0.0, max_value=100.0)
subject2 = st.number_input("Score for Subject 2:", min_value=0.0, max_value=100.0)
subject3 = st.number_input("Score for Subject 3:", min_value=0.0, max_value=100.0)
subject4 = st.number_input("Score for Subject 4:", min_value=0.0, max_value=100.0)
subject5 = st.number_input("Score for Subject 5:", min_value=0.0, max_value=100.0)

if st.button("Calculate"):
    try:
        scores = [subject1, subject2, subject3, subject4, subject5]
        total, average = calculate_scores(scores)
        result = "Pass" if average >= 50 else "Fail"

        st.write(f"Student: {name}")
        st.write(f"Total: {total}")
        st.write(f"Average: {average}")
        st.write(f"Result: {result}")
    except ValueError:
        st.error("Invalid input. Please enter numeric values.")
```
